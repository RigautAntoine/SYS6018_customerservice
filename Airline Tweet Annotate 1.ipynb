```python
import pandas as pd
import numpy as np
# Read in csv file with tweet text for manual annotation
tweets = pd.read_csv("/Users/dylan/msds/data mining/Final Project/SYS6018_customerservice/airlines_set1.csv", encoding = "iso-8859-1")
```

```python
# Create empty column to store label data
tweets["Label"] = np.nan
```

```python
# Create function to take user input and assign it as a label

def tweetLabel(n):
    print(tweets.text[n])
    label = input("Is this a customer complaint? y/n: ")
    if label.lower() == "y" or label.lower() == "n":
        tweets.loc[n,"Label"] = label
    else:
        print("Please enter a valid label (y/n)")
        tweetLabel(n)
```

```python
# Loop through tweets and present them to the user for labeling
for i in range(0,len(tweets.Label)):
    tweetLabel(i)
```
